```python
from unstructured.partition.pdf import partition_pdf

element = partition_pdf(
            filename="MANUAL_SJPH Rev 01 -.pdf",
            strategy="hi_res",
            infer_table_structure=True,
            extract_images_in_pdf=False,
            model_name="yolox",
            chunking_strategy="by_title"
)
```

```python
text = []

for e in element:
    if 'Table' in str(type(e)):
        print(e.metadata.text_as_html)
        text.append(e.metadata.text_as_html)
    else:
        print(e)
        text.append(e.text)
    
```

```python
import chromadb
from config import Config

client = chromadb.PersistentClient(path=Config.CHROMADB_DIRECTORY)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
from langchain_chroma import Chroma

vector_db = Chroma(
    client=client,
    collection_name="ISO",
    embedding_function=embedding
)
```

```python
from langchain_core.documents import Document

documents = [Document(page_content=t, metadata={"source": "MANUAL_SJPH Rev 01 -.pdf"}) for t in text]
```

```python
from uuid import uuid4

ids = [str(uuid4()) for _ in range(len(documents))]

vector_db.add_documents(documents=documents, ids=ids)
```

```python
result = vector_db.similarity_search(
    "jelaskan tetang kebijakan halal",
    k=2
) 
```
